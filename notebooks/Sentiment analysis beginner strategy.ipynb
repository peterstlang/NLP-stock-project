#### Reading data from file


```python
# imports
import pandas as pd
```

```python
# loading data
df = pd.read_csv("C:/Users/peter/Documents/NLP-stock-project/data/news_headlines.csv")
df.head()
```

```python
df.shape
```

```python
#import
from langdetect import detect

# Adding a language column
df["Language"] = df["Title"].apply(lambda x: detect(x))
# Keeping only rows where lang is en
df = df.drop(df[df.Language != "en"].index)
df = df.drop("Language", axis=1)

# clean PublishedAt column
df["Published At"] = pd.to_datetime(df["Published At"], format="%Y-%m-%dT%H:%M:%SZ")

df.head()
```

```python
# WHY IS DATA NOT PULLED AND SAVED?
df.shape
```

```python
for title in df["Title"]:
    print(title)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA

results_headline = []
results_desc = []

for headline, desc in zip(df["Title"], df['Description']):
    # headline
    pol_score_headline = SIA().polarity_scores(headline)  # run analysis
    pol_score_headline['Text'] = headline  # add text for viewing
    results_headline.append(pol_score_headline)

    # description
    pol_score_desc = SIA().polarity_scores(desc)  # run analysis
    pol_score_desc['Text'] = desc  # add text for viewing
    results_desc.append(pol_score_desc)

len(results_headline)

# Print NaN values
for i, score in enumerate(pd.DataFrame(results_headline)['compound']):
    if pd.isna(score):
        print(f"Index: {i}, Headline: {df['Title'][i]}")
```

```python
df = df.reset_index(drop=True)
```

```python
# ADD TO DF
df['Score_headline'] = pd.DataFrame(results_headline)['compound']
df['Score_desc'] = pd.DataFrame(results_desc)['compound']
```
